```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
```

```python
cnn = Sequential()
```

```python
cnn.add(Input(shape=(28,28,1)))
cnn.add(Conv2D(32,(3,3),activation='relu'))
cnn.add(MaxPooling2D(2,2))
cnn.add(Conv2D(64,(3,3),activation='relu'))
cnn.add(MaxPooling2D(2,2))
cnn.add(Conv2D(128,(3,3)))
cnn.add(Flatten())
```

```python
cnn.add(Dense(728,activation='relu'))
cnn.add(Dropout(0.2))
cnn.add(Dense(256,activation='relu'))
cnn.add(Dropout(0.2))
cnn.add(Dense(10,activation='softmax'))
```

```python
cnn.summary()
```

```python
train_datagen=ImageDataGenerator(rescale=1./255)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
train_generator=train_datagen.flow_from_directory('/kaggle/input/datasets/scolianni/mnistasjpg/trainingSet/trainingSet',target_size=(28,28),batch_size=512,class_mode='categorical',color_mode='grayscale')
```

```python
test_generator=test_datagen.flow_from_directory('/kaggle/input/datasets/scolianni/mnistasjpg/trainingSample/trainingSample',target_size=(28,28),batch_size=512, class_mode='categorical',color_mode='grayscale')
```

```python
cnn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(train_generator,epochs=20,validation_data=test_generator)
```

```python
loss, accuracy = cnn.evaluate(test_generator)

print("Accuracy:", accuracy)
```
